# McCulloch-Pitts Neuron Implementation for AND Gate

# Definitions and Formulae:
1. Summation Function: S = Î£ Xi, where Xi are inputs.
2. Threshold Function: Output = 1 if S >= threshold, else 0.
3. Accuracy Score: Accuracy = (Correct Predictions / Total Predictions) * 100.
4. Confusion Matrix: A table showing actual vs. predicted values for evaluation.

Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score
```

Step 1: Define the McCulloch-Pitts Neuron Function

```python
def mcp_neuron(inputs, threshold):
    """
    Implements a simple McCulloch-Pitts neuron.
    :param inputs: Binary input array
    :param threshold: Activation threshold
    :return: Binary output (0 or 1)
    """
    summation = np.sum(inputs)  # Compute summation
    return 1 if summation >= threshold else 0  # Apply threshold
```

Step 2: Define the AND Gate Inputs and Expected Outputs

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # Input combinations
y_true = np.array([0, 0, 0, 1])  # Expected outputs for AND gate
```

Step 3: Define Threshold (Pre-determined for AND Gate)

```python
threshold = 2  # Activation threshold for AND operation
```

Step 4: Compute Model Predictions

```python
y_pred = np.array([mcp_neuron(x, threshold) for x in X])
```

Step 5: Evaluate the Model Performance

```python
accuracy = accuracy_score(y_true, y_pred)
conf_matrix = confusion_matrix(y_true, y_pred)
```

Step 6: Visualize the Decision Boundary

```python
plt.figure(figsize=(6,6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=["0", "1"], yticklabels=["0", "1"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix for McCulloch-Pitts Neuron (AND Gate)")
plt.show()
```

Scatter plot of decision boundary with separating line

```python
# Scatter plot of decision boundary with a separating line
fig, ax = plt.subplots(figsize=(6,6))

# Convert y_true to a NumPy array to allow proper indexing
y_true = np.array(y_true)

# Plot points with explicit labels for legend
ax.scatter(X[y_true == 0][:, 0], X[y_true == 0][:, 1], c='red', edgecolors='black', s=100, label='Class 0 (False)')
ax.scatter(X[y_true == 1][:, 0], X[y_true == 1][:, 1], c='blue', edgecolors='black', s=100, label='Class 1 (True)')
```

Draw decision boundary (x + y = threshold, i.e., x + y = 2)

```python
x_values = np.linspace(-1, 2, 100)  # Generate x-values for plotting the line
y_values = 2 - x_values  # Calculate corresponding y-values for x + y = 2

# Create the plot
plt.figure(figsize=(6, 6))

# Plot the AND Gate outputs: Green for output 1, Red for output 0
plt.scatter(X[y_true == 1][:, 0], X[y_true == 1][:, 1], color='green', label='Output 1', s=100, edgecolors='black')
plt.scatter(X[y_true == 0][:, 0], X[y_true == 0][:, 1], color='red', label='Output 0', s=100, edgecolors='black')

# Plot the decision boundary line: x + y = 2
plt.plot(x_values, y_values, color='blue', label='Decision Boundary: x + y = 2')

# Adding labels and title
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('AND Gate Implementation with Decision Boundary')

# Show legend
plt.legend()

# Display the plot
plt.grid(True)
plt.xlim(-1, 2)
plt.ylim(-1, 2)
plt.show()
```

Print Performance Metrics
