Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

Define input values (X) and their corresponding expected outputs (Y)

```python
X = np.array([0.5, 2.5])
Y = np.array([0.2, 0.9])
```

Sigmoid activation function

Formula: f(w, b, x) = 1 / (1 + exp(-(w*x + b)))

```python
def f(w, b, x):
    return 1.0 / (1.0 + np.exp(-(w * x + b)))
```

Error function (Mean Squared Error)

Formula: E = 1/2 * Σ (f(w,b,x) - y)^2

```python
def error(w, b):
    err = 0.0
    for x, y in zip(X, Y):
        fx = f(w, b, x)
        err += 0.5 * (fx - y) ** 2
    return err
```

Gradient of error w.r.t weight (w)

Formula: ∂E/∂w = (f(w, b, x) - y) * f(w, b, x) * (1 - f(w, b, x)) * x

```python
def grad_w(w, b, x, y):
    fx = f(w, b, x)
    return (fx - y) * fx * (1 - fx) * x
```

Gradient of error w.r.t bias (b)

Formula: ∂E/∂b = (f(w, b, x) - y) * f(w, b, x) * (1 - f(w, b, x))

```python
def grad_b(w, b, x, y):
    fx = f(w, b, x)
    return (fx - y) * fx * (1 - fx)
```

Adagrad Algorithm (Adaptive Learning Rate)

Modifies the learning rate dynamically based on past squared gradients

```python
def do_adagrad():
    w, b = -2, -2  # Initialize weight and bias
    eta = 0.1  # Initial learning rate
    v_w, v_b = 0, 0  # Squared sum of past gradients
    eps = 1e-8  # Smoothing term to prevent division by zero
    max_epochs = 1000  # Number of training iterations
    w_path, b_path = [], []  # To store trajectory

    for i in range(max_epochs):
        dw, db = 0, 0  # Initialize gradients
        for x, y in zip(X, Y):  # Compute gradients
            dw += grad_w(w, b, x, y)
            db += grad_b(w, b, x, y)

        # Accumulate squared gradients (Adagrad update)
        v_w += dw ** 2
        v_b += db ** 2

        # Update parameters (Adagrad formula)
        w = w - (eta / np.sqrt(v_w + eps)) * dw
        b = b - (eta / np.sqrt(v_b + eps)) * db

        # Store values for visualization
        w_path.append(w)
        b_path.append(b)

    return w_path, b_path
```

RMSProp Algorithm (Root Mean Square Propagation)

Instead of accumulating all past squared gradients (like Adagrad), it **decays** past gradients exponentially

```python
def do_rmsprop():
    w, b = -2, -2  # Initialize weight and bias
    eta = 0.1  # Initial learning rate
    v_w, v_b = 0, 0  # Squared moving average of past gradients
    eps = 1e-8  # Smoothing term
    beta1 = 0.9  # Decay factor
    max_epochs = 1000  # Number of training iterations
    w_path, b_path = [], []  # To store trajectory

    for i in range(max_epochs):
        dw, db = 0, 0  # Initialize gradients
        for x, y in zip(X, Y):  # Compute gradients
            dw += grad_w(w, b, x, y)
            db += grad_b(w, b, x, y)

        # Compute moving average of squared gradients (RMSProp update)
        v_w = beta1 * v_w + (1 - beta1) * (dw ** 2)
        v_b = beta1 * v_b + (1 - beta1) * (db ** 2)

        # Update parameters (RMSProp formula)
        w = w - (eta / np.sqrt(v_w + eps)) * dw
        b = b - (eta / np.sqrt(v_b + eps)) * db

        # Store values for visualization
        w_path.append(w)
        b_path.append(b)

    return w_path, b_path
```

Create a mesh grid for 3D visualization

```python
W = np.linspace(-3, 3, 50)  # Range for weight values
B = np.linspace(-3, 3, 50)  # Range for bias values
W, B = np.meshgrid(W, B)  # Create grid
```

Compute the error function on the entire grid

```python
Z = np.array([[error(w, b) for w, b in zip(W_row, B_row)] for W_row, B_row in zip(W, B)])
```

Get the optimization paths using Adagrad and RMSProp

```python
adagrad_w_path, adagrad_b_path = do_adagrad()
rmsprop_w_path, rmsprop_b_path = do_rmsprop()
```

3D Visualization of the error surface and gradient descent paths

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the error surface
ax.plot_surface(W, B, Z, cmap='viridis', alpha=0.8)

# Plot the optimization paths
ax.plot(adagrad_w_path, adagrad_b_path, [error(w, b) for w, b in zip(adagrad_w_path, adagrad_b_path)],
        color='red', label="Adagrad Path")

ax.plot(rmsprop_w_path, rmsprop_b_path, [error(w, b) for w, b in zip(rmsprop_w_path, rmsprop_b_path)],
        color='blue', label="RMSProp Path")

# Label axes
ax.set_xlabel('Weight (w)')
ax.set_ylabel('Bias (b)')
ax.set_zlabel('Error')
ax.set_title('3D Visualization: Adagrad vs RMSProp')

plt.legend()
plt.show()
```
