One-hot encoding is a method used to represent words in a way that machines can understand.

Each unique word in the vocabulary is represented as a binary vector.

If the vocabulary size is N, each word is represented as an N-dimensional vector with a single '1' and the rest '0's.

Import necessary libraries

```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

Sample corpus (list of words)

```python
corpus = ["apple", "banana", "orange", "grape", "apple", "orange"]
```

Reshape data to a 2D array as required by OneHotEncoder


```python
words_array = np.array(corpus).reshape(-1, 1)
```

Initialize OneHotEncoder

```python
encoder = OneHotEncoder(sparse_output=False)  # sparse=False to get dense numpy array output
```

Fit and transform the words to one-hot encoding

```python
one_hot_vectors = encoder.fit_transform(words_array)
```

Display the mapping of words to one-hot vectors

```python
print("Vocabulary:", encoder.categories_)
print("One-hot encoded vectors:")
for word, vector in zip(corpus, one_hot_vectors):
    print(f"{word}: {vector}")
```
